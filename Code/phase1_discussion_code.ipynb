```python
import matplotlib.pyplot as plt

def plot_metrics(b1, b2, b3, b4, m, r_l, cider, spice, spider):
    metrics = ['B-1', 'B-2', 'B-3', 'B-4', 'M', 'R_L', 'Cider', 'Spice', 'Spider']
    resnet50 = [b1[:8], b2[:8], b3[:8], b4[:8], m[:8], r_l[:8], cider[:8], spice[:8], spider[:8]]
    vithuge = [b1[8:], b2[8:], b3[8:], b4[8:], m[8:], r_l[8:], cider[8:], spice[8:], spider[8:]]
    
    # Compute the average for each metric
    resnet50_averages = [np.mean(metric) for metric in resnet50]
    vithuge_averages = [np.mean(metric) for metric in vithuge]

    x = np.arange(len(metrics))
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots(figsize=(12, 7))
    bars1 = ax.bar(x - width/2, resnet50_averages, width, label='ResNet-50')
    bars2 = ax.bar(x + width/2, vithuge_averages, width, label='ViT_huge')

    ax.set_xlabel('Metrics')
    ax.set_ylabel('Average Score')
    ax.set_title('Average Scores per Metric between CLIP encoders')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()
    ax.grid(True)

    plt.show()
```
