```python
def tokenization(my_dataset, tokenizer, synthetic_caps=True):
    my_tokenized_data = {}
    # The text is in the form of (captions,references) 
    for key, text in list(my_dataset.items())[:-1]:
        captions = text[0]
        references = text[1]
        if synthetic_caps:
            tokenized_captions = [tokenizer(sub_captions, padding=True, truncation=True, return_tensors='pt') for sub_captions in captions]
        else: 
            tokenized_captions = tokenizer(captions, padding=True, truncation=True, return_tensors='pt')
        tokenized_references = tokenizer(references, padding=True, truncation=True, return_tensors='pt')
        my_tokenized_data[key] = (tokenized_captions, tokenized_references)

    return my_tokenized_data
```

```python
def transform_pad_entry(tokenized_entry):
    padded_entry = {}
    def my_transpose_pad(list_of_tensors):
        transposed_tensors = [tensor.t() for tensor in list_of_tensors]
        padded_tensors = pad_sequence(transposed_tensors, batch_first=True, padding_value=0)
        padded_tensors = [tensor.t() for tensor in padded_tensors]
        return padded_tensors

    # ------------------------ Pad the input-ids from the entry data -----------------------------
    temp_input_ids = [tokenized_entry[i]['input_ids'] for i in range(5)]
    padded_input_ids = my_transpose_pad(temp_input_ids)
    padded_entry['input_ids'] = padded_input_ids

    # ------------------------ Pad the token-type-ids from the entry data ---------------------------
    temp_token_type_ids = [tokenized_entry[i]['token_type_ids'] for i in range(5)]
    padded_token_type_ids = my_transpose_pad(temp_token_type_ids)
    padded_entry['token_type_ids'] = padded_token_type_ids

    # ------------------------ Pad the attention-mask from the entry data ---------------------------
    temp_attention_mask = [tokenized_entry[i]['attention_mask'] for i in range(5)]
    padded_att_mask = my_transpose_pad(temp_attention_mask)
    padded_entry['attention_mask'] = padded_att_mask

    return(padded_entry)
```

## Setting the T-to-T Transformer

```python
import pickle, torch
from torch.utils.data import DataLoader, Dataset
import torch.nn.functional as F
from torch.nn.utils.rnn import pad_sequence

class CustomDataset(Dataset):
    def __init__(self, entries, references):
        self.entries = entries
        self.references = references

    def __len__(self):
        return len(self.entries)

    def __getitem__(self, idx):
        entry = self.entries[idx]
        reference = self.references[idx]
        return entry, reference


def custom_collate_fn(batch):
    entries, references = zip(*batch)
    
    # Determine the max length in the batch for entries and references
    max_len_entry = max(entry.size(-1) for entry in entries)
    max_len_ref = max(ref.size(-1) for ref in references)
    
    # Pad entries
    padded_entries = [F.pad(entry, (0, max_len_entry - entry.size(-1))) for entry in entries]
    padded_entries = torch.stack(padded_entries, dim=0)
    
    # Pad references
    padded_references = [F.pad(ref, (0, max_len_ref - ref.size(-1))) for ref in references]
    padded_references = torch.stack(padded_references, dim=0)
    
    return padded_entries, padded_references
```

```python
import torch.nn.functional as F
import re

def generate_text_from_logits(logits, tokenizer, strategy='greedy'):
    """
    Generate text from logits using the specified strategy.

    Parameters:
    logits (torch.Tensor): The logits output from the model of shape [batch_size, num_sentences, target_seq_len, vocab_size].
    tokenizer (object): The tokenizer or vocabulary object to convert tokens to words.
    strategy (str): The decoding strategy, 'greedy' or 'sampling'.

    Returns:
    List[List[str]]: The generated text for each sentence.
    """
    if len(logits.shape)==3:
        logits = logits.unsqueeze(0)
    else:
        pass

    all_stories_in_btch = []
    for j in range(logits.shape[0]):
        generated_sentences = []
        # Iterate over each sentence
        for sentence_logits in logits[j,:,:,:]:
            sentence_tokens = []

            # Iterate over each token in the sentence
            for token_logits in sentence_logits:
                if strategy == 'greedy':
                    # Greedy decoding: select the token with the highest probability
                    token_id = torch.argmax(F.softmax(token_logits, dim=-1)).item()
                elif strategy == 'sampling':
                    # Sampling decoding: sample a token based on the probability distribution
                    probabilities = F.softmax(token_logits, dim=-1)
                    token_id = torch.multinomial(probabilities, 1).item()
                else:
                    raise ValueError("Unsupported decoding strategy")

                # Convert token ID to word using the tokenizer or vocabulary
                token_word = tokenizer.decode([token_id])
                sentence_tokens.append(token_word)

            # Join the tokens to form the sentence
            generated_sentence = ' '.join(sentence_tokens)
            generated_sentences.append(generated_sentence)


        all_stories_in_btch.append(generated_sentences)

    return all_stories_in_btch

def merge_subword_tokens(tokens):
    merged_tokens = []
    for token in tokens:
        if token.startswith('##'):
            if merged_tokens:
                merged_tokens[-1] += token[2:]
            else:
                merged_tokens.append(token[2:])
        else:
            merged_tokens.append(token)
    return merged_tokens


# Refine the sentences before printing. Remove pad, ending and beginning tokens 
def refine_results(sentences, nam):
    tokens_to_remove = ['[CLS]', '[SEP]', '[PAD]', '[UNK]']
    #tokens_to_remove = []
    merged_sentences = [sentence.split() for sentence in sentences]
    get_index_of_end_tok = [sentence.index('[SEP]') for sentence in merged_sentences]
    chopped_sentences = [[token for token in sentence[:idx+1]] for idx,sentence in zip(get_index_of_end_tok,merged_sentences)]
    refined_sentences =  [' '.join([token for token in sentence if token not in tokens_to_remove]) for sentence in chopped_sentences]
    story_sentences = [re.sub(r'\s([?.!,\'](?:\s|$))', r'\1', sentence) for sentence in refined_sentences]
    final_story = ' '.join(story_sentences)
    if nam=='admitos':
        split_story = final_story.split()
        merged_story = merge_subword_tokens(split_story)
        merge_story_sent = []
        for sent in story_sentences:
            merged_sent = merge_subword_tokens(sent.split())
            merge_story_sent.append(' '.join(merged_sent))
        return ' '.join(merged_story), merge_story_sent
    if nam=='yingjin':
        split_story = final_story.split()
        merged_story = merge_subword_tokens(split_story)
        return ' '.join(merged_story)


# Generate text using greedy decoding
all_final_stories = []
for i in range(len(all_outputs[:1])):
    print("Generation:", i)
    generated_stories = generate_text_from_logits(all_outputs[i], tokenizer, strategy='greedy') 
    if name == 'admitos':
        batch_final_stories = [refine_results(story,name) for story in generated_stories]
        all_final_stories.append(batch_final_stories)
    if name == 'yingjin':
        batch_final_stories = [refine_results(story,name) for story in generated_stories]
        all_final_stories.append(batch_final_stories)
    
```

```python
def encode_data(data, tokenizer, max_length=512):
    input_encodings = [tokenizer(item[0], padding=True, truncation=True, max_length=max_length, return_tensors="pt") for item in data.values()]
    target_encodings = [tokenizer(item[1], padding=True, truncation=True, max_length=max_length, return_tensors="pt") for item in data.values()]
    return input_encodings, target_encodings
```

```python
class CustomDatasetT5(Dataset):
    def __init__(self, input_data, target_data):
        self.input_data = input_data
        self.target_data = target_data

    def __len__(self):
        return len(self.input_data)

    def __getitem__(self, idx):
        input_encodings = self.input_data[idx]
        target_encodings = self.target_data[idx]

        input_ids = input_encodings['input_ids'].clone().detach()
        attention_mask = input_encodings['attention_mask'].clone().detach()

        target_ids = target_encodings['input_ids'].clone().detach()
        target_attention_mask = target_encodings['attention_mask'].clone().detach()

        return {'input_ids': input_ids, 'attention_mask': attention_mask, 'target_ids': target_ids, 'target_attention_mask': target_attention_mask}
    
def my_collate_fn(batch):
    input_ids = [item['input_ids'].transpose(0,1) for item in batch]
    attention_mask = [item['attention_mask'].transpose(0,1) for item in batch]
    target_ids = [item['target_ids'].transpose(0,1) for item in batch]
    target_attention_mask = [item['target_attention_mask'].transpose(0,1) for item in batch]
    
    input_ids_padded = pad_sequence(input_ids, batch_first=True, padding_value=0)
    attention_mask_padded = pad_sequence(attention_mask, batch_first=True, padding_value=0)
    target_ids_padded = pad_sequence(target_ids, batch_first=True, padding_value=0)
    target_attention_mask_padded = pad_sequence(target_attention_mask, batch_first=True, padding_value=0)

    final_input_ids = [item.transpose(0,1) for item in input_ids_padded]
    final_attention_masks = [item.transpose(0,1) for item in attention_mask_padded]
    final_target_ids = [item.transpose(0,1) for item in target_ids_padded]
    final_target_attention_masks = [item.transpose(0,1) for item in target_attention_mask_padded]

    return {'input_ids': final_input_ids, 'attention_mask': final_attention_masks, 'target_ids': final_target_ids, 'target_attention_mask': final_target_attention_masks}
```

```python
def prepare_data(data):
    inputs = []
    outputs = []
    for captions, story in data:
        input_text = ' [SEP] '.join(captions)
        inputs.append(input_text)
        outputs.append(story)
    return inputs, outputs

t5_train_data = [prepare_data([train_tup]) for train_tup in list(new_train_dataset.values())]
t5_val_data = [prepare_data([val_tup]) for val_tup in list(new_val_dataset.values())]
t5_test_data = [prepare_data([test_tup]) for test_tup in list(new_test_dataset.values())]

t5_train_caps =  [item[0] for item in t5_train_data]
t5_train_stories =  [item[1] for item in t5_train_data]

t5_val_caps =  [item[0] for item in t5_val_data]
t5_val_stories =  [item[1] for item in t5_val_data]

t5_test_caps =  [item[0] for item in t5_test_data]
t5_test_stories =  [item[1] for item in t5_test_data]
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
tokenizer = T5Tokenizer.from_pretrained('t5-small')

train_caps_encodings = [tokenizer(train_cap, padding=True, truncation=True, return_tensors="pt") for train_cap in t5_train_caps]
train_story_labels = [tokenizer(train_story, padding=True, truncation=True, return_tensors="pt").input_ids for train_story in t5_train_stories]

val_caps_encodings = [tokenizer(val_cap, padding=True, truncation=True, return_tensors="pt") for val_cap in t5_val_caps]
val_story_labels = [tokenizer(val_story, padding=True, truncation=True, return_tensors="pt").input_ids for val_story in t5_val_stories]

test_caps_encodings = [tokenizer(test_cap, padding=True, truncation=True, return_tensors="pt") for test_cap in t5_test_caps]
test_story_labels = [tokenizer(test_story, padding=True, truncation=True, return_tensors="pt").input_ids for test_story in t5_test_stories]
```

```python
from torch.utils.data import DataLoader, Dataset 
from torch.nn.utils.rnn import pad_sequence

class StoryDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: val.clone().detach() for key, val in self.encodings[idx].items()}
        item['labels'] = self.labels[idx].clone().detach()
        return item

    def __len__(self):
        return len(self.labels)
    
def my_collate_fn(batch):
    input_ids = [item['input_ids'].transpose(0,1) for item in batch]
    attention_mask = [item['attention_mask'].transpose(0,1) for item in batch]
    labels = [item['labels'].transpose(0,1) for item in batch]
    
    input_ids_padded = pad_sequence(input_ids, batch_first=True, padding_value=0)
    attention_mask_padded = pad_sequence(attention_mask, batch_first=True, padding_value=0)
    labels_padded = pad_sequence(labels, batch_first=True, padding_value=0)

    final_input_ids = [item.transpose(0,1) for item in input_ids_padded]
    final_attention_masks = [item.transpose(0,1) for item in attention_mask_padded]
    final_labels = [item.transpose(0,1) for item in labels_padded]

    return {'input_ids': final_input_ids, 'attention_mask': final_attention_masks, 'labels': final_labels}



train_dataset = StoryDataset(train_caps_encodings, train_story_labels)
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=my_collate_fn)

val_dataset = StoryDataset(val_caps_encodings, val_story_labels)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=True, collate_fn=my_collate_fn)

test_dataset = StoryDataset(test_caps_encodings, test_story_labels)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=True, collate_fn=my_collate_fn)

```

```python
# Initialize the model
model = T5ForConditionalGeneration.from_pretrained('t5-small')
```

```python
# Define optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

# Training loop
num_epochs = 3
model.train()
for epoch in range(num_epochs):
    train_loss = 0.0
    for it, batch in enumerate(train_loader):
        optimizer.zero_grad()
        batch_input_ids = torch.stack(batch['input_ids']).squeeze(1)
        batch_att_masks = torch.stack(batch['attention_mask']).squeeze(1)
        batch_labels = torch.stack(batch['labels']).squeeze(1)
        final_labels = batch_labels.clone()
        final_labels[batch_labels == tokenizer.pad_token_id] = -100

        outputs = model(input_ids=batch_input_ids, attention_mask=batch_att_masks, labels=final_labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        print(f'Train Iteration {it} completed with running loss: {loss.item()}')
        train_loss += loss.item()
        if it>0:
            break
        
    print(f'Epoch {epoch} completed with train loss: {train_loss/(it+1)}')

    model.eval()
    val_loss = 0
    with torch.no_grad():
        for it, batch in enumerate(val_loader):
            batch_input_ids = torch.stack(batch['input_ids']).squeeze(1)
            batch_att_masks = torch.stack(batch['attention_mask']).squeeze(1)
            batch_labels = torch.stack(batch['labels']).squeeze(1)
            final_labels = batch_labels.clone()
            final_labels[batch_labels == tokenizer.pad_token_id] = -100

            outputs = model(input_ids=batch_input_ids, attention_mask=batch_att_masks, labels=final_labels)
            val_loss += outputs.loss.item()
            print(f'Val Iteration {it} completed with running loss: {loss.item()}')
            if(it>2):
                break

    print(f"Validation Loss: {val_loss/(it+1)}")
    break 
```
