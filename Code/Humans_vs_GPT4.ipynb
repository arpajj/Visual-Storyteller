```python
import matplotlib.pyplot as plt
import numpy as np

# Transition scores for each model
D1 = [6,8,6,2,5,7,7,8,8,7,8,8,8,6,5,5,5,5,6,4,7,6,2,6]
D2 = [6,9,6,3,4,6,8,7,7,7,6,7,6,7,7,7,3,2,10,8,10,10,10,8]
D3 = [4,6,8,1,8,5,5,6,7,8,8,7,8,8,6,7,9,6,7,7,3,5,5,5]
D4 = [10,7,10,3,7,6,7,8,6,6,7,7,6,8,8,7,8,10,10,10,7,10,8,10]
HMM = [5,7,5,5,6,5,6,7,7,8,8,8,3,5,6,10,8,9,10,10,10,10,10,9]

# Data for the boxplot
data2 = [D1, D2, D3, D4, HMM]

# Labels for the models
labels = ['F1', 'F2', 'F3', 'F4', 'Humans']

# Create the boxplot with custom colors
plt.figure(figsize=(10, 6))
#box = plt.boxplot(data, labels=labels, patch_artist=True)
box = plt.boxplot(data2, labels=labels, patch_artist=True, medianprops=dict(color='black', linewidth=2))
means = [np.mean(D1), np.mean(D2), np.mean(D3), np.mean(D4), np.mean(HMM)]
print(means)
my_color = 'yellow'

# Customize box colors to red
for patch in box['boxes']:
    patch.set(facecolor='red')

# Customize the median line color to purple
for median in box['medians']:
    median.set(color=my_color, linewidth=2)

# Customize the mean markers to be blue
plt.scatter(range(1, 6), means, color='blue', marker='o', s=25, zorder=2, label='Mean')
plt.plot([], [], color=my_color, linewidth=2, label='Median')


# Add titles and labels
plt.title('Distribution of Human Transition Scores for each Framework', fontsize=15)
plt.ylabel('$Transition \quad Scores$', fontsize = 13)
plt.legend(loc='best')
plt.xlabel("$Tested \quad Frameworks$", fontsize = 13)
plt.yticks(np.arange(0,11,1))


# Set plot limits and grid
plt.ylim(0, 11)
plt.grid(True)

# Show plot
plt.show()
```
